# Pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/master/Iris.csv')

df.drop('Id', axis=1, inplace = True)

df.head()
```

```python
# Plot all the numrical columns of a table

df.plot(figsize  = (10,3))
```

```python
# Draw a Bar chart

print(df.iloc[5:11])
df.iloc[5:11].plot(kind = 'bar',figsize = (10,3))
```

```python
# Horizontal Bar Chart

print(df.iloc[[5]])
df.iloc[[5]].plot(kind = 'barh',figsize = (10,3))
```

```python
# Histogram

df.plot(kind = 'hist')
```

```python
df['SepalLengthCm'].plot(kind = 'hist',orientation = 'horizontal')
```

```python
df.hist(figsize=(20,10),color = 'g' , alpha = .9)
```

```python
# Box Plot

df.plot(kind = 'box' ,figsize = (10,3),color = {'boxes' : 'g' , 'whiskers' :'r'} , vert = False )
```

```python
# Area Chart

df.plot(kind = 'area' , figsize = (10,3), alpha = .4,stacked = False)
```

```python
# Scatter Plot 

df.plot.scatter(x = 'SepalLengthCm' , y = 'PetalLengthCm', figsize = (10,3))
```

```python
# Bubble Chart

df.plot.scatter(x = 'SepalLengthCm' , y = 'PetalLengthCm',c = 'SepalWidthCm',s = df['PetalWidthCm']*500, figsize = (10,3))
```

```python
# Pie Chart
```

```python
df1 = df.drop(columns='Species').iloc[0]
df1.head()
```

```python
df1.plot(kind = 'pie', figsize = (10,3))
```

# Matplotlib

```python
# Load the library

import matplotlib.pyplot as plt
```

```python
# initializing the data
x = [10, 20, 30, 40]
y = [20, 30, 40, 50]
  
# plotting the data
plt.plot(x, y)
  
# Adding the title
plt.title("Simple Plot")
  
# Adding the labels
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
# Giving Marker points in chart

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# Set line style

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
# Create labels for a plot

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
# Add Grid Lines to a Plot

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
# Display Multiple Plots

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
# Scatter Plot

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
# Bar Chart

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
# Histogram

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

```python
# Pie Chart

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

# Seaborn

```python
# Import the library

import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.scatterplot(x = df.sepal_length , y = df.petal_length,)
```

```python
# Distribution Plot

sns.distplot(df['sepal_length'])
```

```python
# Load tips dataset
df1 = sns.load_dataset('tips')
df1.head()
```

```python
# Relational Plot or Advanced Scatter Plot
sns.relplot(x = 'total_bill', y = 'tip',data = df1)
```

```python
sns.relplot(x = 'total_bill' , y ='tip'  , data = df1 ,hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill' , y ='tip'  , data = df1 ,style = 'time',hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill' , y ='tip' , kind  = 'line' , data = df1)
```

```python
sns.relplot(x = 'total_bill' , y ='tip' , col= 'time',data =df1)
```

```python
sns.relplot(x='total_bill', y='tip',data=df1,col='day')
```

```python
# Categorical Plot
sns.catplot(x = 'day' , y = 'total_bill', data = df1)
```

```python
sns.catplot(x = 'day' , y = 'tip',kind = 'box' , data = df1,hue = 'sex')
```

```python
# To plot multiple pairwise bivariate distributions in a dataset
sns.pairplot(df)
```
